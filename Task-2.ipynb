```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
```

```python
print("Summary Statistics")
print(df.describe())
print("\nMissing Values")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(10,5))
df['Age'].hist(bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
px.histogram(df, x="Age", title="Interactive Age Distribution").show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title("Fare by Passenger Class")
plt.show()
```

```python
px.box(df, x="Pclass", y="Fare", title="Interactive Boxplot of Fare by Passenger Class").show()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']].dropna(), hue='Survived')
plt.show()
```

```python
px.scatter(df, x="Age", y="Fare", color="Survived", title="Interactive Scatterplot: Age vs. Fare").show()
```
